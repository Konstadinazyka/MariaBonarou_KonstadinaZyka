ΠΕΡΙΛΗΨΗ ΠΡΟΤΖΕΚΤ:
Δεδομένου ότι η πράσινη ανάπτυξη είναι ένας όρος που προοδευτικά μπαίνει στις ζωές μας, επιλέξαμε να μιλήσουμε για την πορεία των Βαλκανικών χωρών, συμπεριλαμβανομένου και της Ελλαδάς, προς την πράσινη ανάπτυξη αναλύοντας δύο πολύ βασικά συστατικά προς ένα βιωσιμότερο μέλλον, που είναι τα ποσοστά εκπομπών αερίων του θερμοκηπίου και τα ποσοστά αξιοποίησης ανανεώσιμων πηγών ενέργειας. Τα δεδομένα που θα αξιοποιήσουμε, τα αντήσαμε από τα διαθέσιμα δεδομένα της μη-κερδοσκοπικής οργάνωσης "Global Change Data Lab" και αφορούν τα έτη από το 1850 μέχει και το 2021. Ωστόσο, θα χρησιμοποιήσουμε μόνο τα δεδομένα που αφορούν τα έτη από το 2015, που  υπογράφηκαν και καθιερώθηκαν από τον ΟΗΕ οι στόχοι της βιώσιμης ανάπτυξης, μέχρι και το 2021, για να διαπιστώσουμε κατά πόσο οι χώρες της Βαλκανικής χερσονήσου προσπαθούν να συμβάλλουν στο GreenDeal.
