```python
#imports
import pandas as pd

#set ups

#display all rows of any dfs in this notebook
pd.set_option('display.max_rows',1000)

#display all columns of any dfs in this notebook
pd.set_option('display.max_columns',100)
pd.set_option('display.float_format', lambda x: '%.1f' % x)
import matplotlib.pyplot as plt
```

```python
#read a csv dataset
df=pd.read_csv("EMISSIONS_ANNUAL_1830-2021.csv")
```

```python
#convert the dataset to dataframe 
df=pd.DataFrame(df)
df
```

```python
# seperated the list of values to different columns
#i added one more Column with the name "Empty"
# we have to throw the columns we dont need after we check that we *actually* don't need them
df[['CNTR_NAME', 'ISO3', 'Gas', 'Component', 'Year', 'Data', 'Unit','Empty']] = df['CNTR_NAME,"ISO3","Gas","Component","Year","Data","Unit"'].str.split(",", expand=True)
```

```python
df.head()
```

```python
df.shape
```

```python
#From our dataset we want to work with Balkan countries
#Due to the geopolitical situation in Balkans we wanted to check if there were datas from certain countries
```

```python
dfkosovo = df[(df.CNTR_NAME.str.contains('Kosovo',na=False,case=False))]
dfkosovo.head()
```

```python
df[(df.CNTR_NAME.str.contains('Montenegro',na=False,case=False))]
```

```python
#We created a new dataframe which contains only Balkan contries
balkans = df[df["CNTR_NAME"].isin(['Greece', 'Albania', 'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Kosovo', 'Montenegro', 'North Macedonia', 'Romania', 'Serbia','Slovenia'])]
balkans
```

```python
#Check the types 
balkans.dtypes
```

```python
#Kosovo's datas started from 2008, before that year Kosovo was part of Serbia, we wanted to check if there was animportant difference between 2007 & 2008 in Serbia's datas
dfserbia=balkans[(balkans.CNTR_NAME=='Serbia')&((balkans.Year=='2007')|(balkans.Year=='2008'))]
dfserbia
```

```python
dfserbia.dtypes
```

```python
#We created a new, more clean and specific dataframe witch contained only those collumns 
blk = balkans [['CNTR_NAME','ISO3','Gas','Component','Year','Data']].copy()
blk
```

```python
blk.head()
```

```python
#In order to continue with the analysis we had to remove the Quotation marks from every column and row 
```

```python
blk['ISO3'] = blk['ISO3'].str.replace(r'\W', '', regex=True)
blk
```

```python
blk['Gas'] = blk['Gas'].str.replace(r'\W', '', regex=True)
blk
```

```python
blk['Component'] = blk['Component'].str.replace(r'\W', '', regex=True)
blk
```

```python
blk['Data'] = blk['Data'].str.replace(r'\W', '', regex=True)
blk
```

```python
#Change "Year" from object to intiger
blk['Year']=blk['Year'].astype(int)
```

```python
#Change "Data" from object to float
blk['Data'] = blk['Data'].astype(float)
```

```python
blk.dtypes
```

```python
#From the blk dataframe in "Component" collumn we wanted to keep only the "Total" rows
blk1 = blk[blk.Component=='Total']
blk1
```

```python
# blk1['Year'] = blk1['Year'].astype(str)
```

```python
#We were interested only in datas started from Year 2002, so we drop every year that was smaller in number than 2001
#After that we numbered again our data 
blk1 = blk1[blk1.Year>2001]
blk1= blk1.reset_index().drop(['index',],axis=1)
blk1
```

```python
blk1.to_csv(r"C:\Users\Zikonna\Documents\balkans.csv")
```

```python
#We splitted our dataframe based on the gas 
blkCH4 = blk1[blk1.Gas=='CH4']
blkCH4= blkCH4.reset_index().drop(['index'],axis=1)
blkCH4
```

```python
#We changed the format of the dataframe, so it could be easier for us to work with
CH4= blkCH4.groupby(['CNTR_NAME','Year']).Data.mean()

CH4 = pd.DataFrame(CH4)
CH4= CH4.unstack().T.reset_index()
CH4 = CH4.drop(['level_0'],axis=1)
CH4 = CH4.rename(columns={'Bosnia and Herzegovina':'Bosniaherzegovina',
                         'North Macedonia':'Northmacedonia'})
CH4
```

```python
#We wanted to get the sum of the total every 5 yeasrs for each country
CH4['Albania_avg'] = CH4.Albania.rolling(window=5).mean()
CH4['Bulgaria_avg'] = CH4.Bulgaria.rolling(window=5).mean()
CH4['Bosniaherzegovina_avg'] = CH4.Bosniaherzegovina.rolling(window=5).mean()
CH4['Croatia_avg'] = CH4.Croatia.rolling(window=5).mean()
CH4['Greece_avg'] = CH4.Greece.rolling(window=5).mean()
CH4['Montenegro_avg'] = CH4.Montenegro.rolling(window=5).mean()
CH4['Croatia_avg'] = CH4.Croatia.rolling(window=5).mean()
CH4['NorthMacedonia_avg'] = CH4.Northmacedonia.rolling(window=5).mean()
CH4['Romania_avg'] = CH4.Romania.rolling(window=5).mean()
CH4['Serbia_avg'] = CH4.Serbia.rolling(window=5).mean()
CH4['Slovenia_avg'] = CH4.Slovenia.rolling(window=5).mean()
CH4
CH4.head()
```

```python
#We dropped the country names as we left only the country_averange and we left only the years that had the sum 
CH4=CH4.drop(['Albania','Bulgaria','Bosniaherzegovina','Croatia','Greece','Montenegro','Northmacedonia','Romania','Serbia','Slovenia'], axis=1)
CH4 = CH4.drop(labels=[0,1,2,3,5,6,7,8,10,11,12,13,15,16,17,18], axis=0)
CH4= CH4.reset_index().drop(['index',],axis=1)
CH4
```
