```python
import pandas as pd
#display all rows of any dfs in this notebook
pd.set_option('display.max_rows', None)

#display all columns of any dfs in this notebook
pd.set_option('display.max_columns', None)
```

```python
#importing the dataset
file_path = 'Panel_format_CSV1'
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
#a list of countries I want to keep
countries_to_keep = ['Greece', 'Romania', 'Bulgaria','Croatia','North Macedonia','Serbia', 'Slovenia','Other Europe' ]
```

```python
# Using boolean indexing to filter rows for the specified countries
filtered_df = df[df['Country'].isin(countries_to_keep)]
```

```python
# Display the resulting DataFrame
filtered_df
```

```python
#List of columns I want to keep
columns_to_keep = ['Country', 'Year', 'ISO3166_alpha3','Region','ren_power_ej',	'ren_power_twh',	'renewables_ej',]
```

```python
#Using boolean indexing to filter rows for the specified countries
filtered_df = df[df['Country'].isin(countries_to_keep)]
```

```python
#Keep only the specified columns
filtered_df = filtered_df[columns_to_keep]
```

```python
#Display the resulting DataFrame
filtered_df
```

```python
#List of years I want to keep
years_to_keep = [2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
```

```python
filtered_df = df[df['Country'].isin(countries_to_keep)]
#Using boolean indexing to filter rows for the specified years
filtered_df = filtered_df[filtered_df['Year'].isin(years_to_keep)]
```

```python
#Keeping only the specified columns
filtered_df = filtered_df[columns_to_keep]
```

```python
filtered_df
```

```python
#Reseting the index to get a new numbering column
filtered_df = filtered_df.reset_index(drop=True)
```

```python
filtered_df
```

```python
#Conversion factor from exajoules to terawatt-hours
conversion_factor_ej_to_twh = 278
```

```python
#Choosing the column I want to convert
filtered_df['renewables_TWh'] = filtered_df['renewables_ej'] * conversion_factor_ej_to_twh
```

```python
filtered_df
```

```python
filtered_df.to_csv('filtered_output1.csv', index=False)
```

```python
#I want to drop two more columns-creating a list
additional_columns_to_drop = ['ren_power_ej', 'renewables_ej']
```

```python
# Droping the additional columns
filtered_df = filtered_df.drop(columns=additional_columns_to_drop)
```

```python
filtered_df
```

```python
# Changing the variables position
final = filtered_df.groupby(['Country','Year']).renewables_TWh.mean()

final = pd.DataFrame(final)
final = final.unstack().T.reset_index()
final = final.drop(['level_0'],axis=1)
final
```

```python
# Renaming columns and showing the first few rows
final = final.rename(columns={'Other Europe': 'Othereurope',
                         'North Macedonia':'Northmacedonia'})
final.head()
```

```python
#Calculating rolling averages for specific columns in the DataFrame 
final['Othereurope_avg']=final.Othereurope.rolling(window=5).mean()
final['Bulgaria_avg'] = final.Bulgaria.rolling(window=5).mean()
final['Croatia_avg'] = final.Croatia.rolling(window=5).mean()
final['Greece_avg'] = final.Greece.rolling(window=5).mean()
final['Croatia_avg'] = final.Croatia.rolling(window=5).mean()
final['NorthMacedonia_avg'] = final.Northmacedonia.rolling(window=5).mean()
final['Romania_avg'] = final.Romania.rolling(window=5).mean()
final['Serbia_avg'] = final.Serbia.rolling(window=5).mean()
final['Slovenia_avg'] = final.Slovenia.rolling(window=5).mean()
final
```

```python
#Calculating the average every 5 years
final=final.drop(['Bulgaria','Croatia','Greece','Othereurope','Northmacedonia','Romania','Serbia','Slovenia'], axis=1)
final = final.drop(labels=[0,1,2,3,5,6,7,8,10,11,12,13,15,16,17,18], axis=0)
final= final.reset_index().drop(['index',],axis=1)
final
```
