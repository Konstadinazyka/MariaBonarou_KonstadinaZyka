Η Ευρώπη έχει κάνει ξεκάθαρη την θέση της όσον αφορά την κλιματική κρίση και την λήψη μέτρων προκειμένου να πετύχει τον στόχο της κλιματικής ουδετερότητας μέχρι το 2050. Τα βασικότερα μέτρα που έχουν ληφθεί για την επίτευξη αυτού είναι οι 17 στόχοι της βιωσιμότητας που καθιερώθηκαν το 2015 από τον ΟΗΕ και υπογράφηκαν από τα κράτη μέλη της Ε.Ε., η «Ευρωπαϊκή Πράσινη Συμφωνία» για το 2030 και μετέπειτα για το 2050.


**Τα μέτρα της Ευρώπης για το περιβάλλον**

Οι 17 στόχοι της βιωσιμότητας έχουν ως απώτερο σκοπό τη βελτίωση των συνθηκών διαβίωσης με κύριους πυλώνες το περιβάλλον, την οικονομία και τη διακυβέρνηση. Αυτή η συμφωνία είναι μία ευκαιρία συνεργασίας παλαιών και νέων γενεών στον δρόμο προς τη διασφάλιση ενός κατάλληλου για όλους πλανήτη. Βλέπουμε ότι ο όρος βιωσιμότητα δεν περιορίζεται μόνο στον τομέα του περιβάλλοντος όπως στις απαρχές του, αλλά πλέον περιλαμβάνει και άλλους τομείς που επηρεάζουν τη ζωή του ανθρώπου. Ωστόσο, φαίνεται ότι ακόμα και τώρα στον πυρήνα του όρου βιωσιμότητα το περιβάλλον διαδραματίζει σημαντικό ρόλο, καθώς από τη διατήρηση του περιβάλλοντος εξαρτώνται η βιωσιμότητα της οικονομίας και της διακυβέρνησης. Ο ορισμός της βιωσιμότητας δεν είναι σταθερός, καθώς το περιεχόμενο του όρου αλλάζει ανάλογα με τις ανάγκες κάθε εποχής. Για παράδειγμα οι 17 στόχοι που θεσπίστηκαν το 2015 αντανακλούν τις ανάγκες της δικής μας εποχής :
1. Μηδενική φτώχεια
2. Μηδενική πείνα
3. Καλή υγεία και ευημερία
4. Ποιοτική εκπαίδευση
5. Ισότητα των φίλων
6. Καθαρό νερό και αποχέτευση
7. Φτηνή και καθαρή ενέργεια
8. Αξιοπρεπής εργασία και οικονομική ανάπτυξη
9. Βιομηχανία, καινοτομία και υποδομές
10. Λιγότερες ανισότητες
11. Βιώσιμες πόλεις και κοινότητες
12. Υπεύθυνη κατανάλωση και παραγωγή
13. Δράση για το κλίμα
14. Ζωή στο νερό
15. Ζωή στην στεριά
16. Ειρήνη, δικαιοσύνη και ισχυροί δεσμοί
17. Συνεργασία για τους στόχους


Με τη σειρά της η «Ευρωπαϊκή Πράσινη Συμφωνία» θεσπίστηκε έχοντας στον πυρήνα της ξεκάθαρα την περιβαλλοντική βιωσιμότητα. Στόχος της «Πράσινης Συμφωνίας» για το 2050 είναι η επίτευξη της κλιματικής ουδετερότητας στην Ευρώπη. Καθώς όμως, ο στόχος αυτός είναι πολύ μακρινός και οι επιπτώσεις της κλιματικής κρίσης όχι μόνο παραμένουν αισθητές αλλά και γίνονται εντονότερες, προτάθηκε από την Ευρωπαϊκή Επιτροπή η θέσπιση κάποιων ενδιάμεσων στόχων οι οποίοι εντάχθηκαν στην Πράσινη Συμφωνία για το 2030. Αυτή συμφωνία προσκαλεί ουσιαστικά τα κράτη-μέλη στην επίτευξη των ενδιάμεσων αυτών στόχων, ως δείγμα προόδου μέσα σε ένα μικρότερο χρονικό πλαίσιο. Η Ευρωπαϊκή Επιτροπή στοχεύει, μέσω αυτής της τακτικής, να προκαλέσει τα κράτη-μέλη σε μία αναθεώρηση των μέτρων και των σχεδίων που έχουν ληφθεί από μέρους τους. Με αυτόν τον τρόπο διενεργείται ένας έλεγχος ως προς την επίτευξη των στόχων.  

**Η θέση των Βαλκανικών χωρών σε αυτά τα μέτρα**

Η “Ευρωπαϊκη Πράσινη Συμφωνία” θέτει συγκεκριμένους στόχους, σχέδια αλλά και απαιτήσεις προς τις χώρες της Ευρώπης στον δρόμο προς την κλιματική ουδετερότητα. Τα δεδομένα που αναλύονται παρακάτω αφορούν τις τιμές εκπομπών αερίων του θερμοκηπίου και τις τιμές κατανάλωσης πράσινης ενέργειας στις χώρες των Βαλκανίων σε μία προσπάθεια να δούμε με νούμερα κατά πόσο αυτές οι χώρες παρουσιάζουν πρόοδο προς την επίτευξη των στόχων την “Πράσινης Συμφωνίας”. Η επιλογή να επικεντρωθεί η έρευνα σε αυτήν την συγκεκριμένη γεωγραφική περιοχή έγινε για να περιοριστεί και να εξειδικευτεί η έρευνα. Στόχος της έρευνας είναι να δούμε κατά πόσο τα ιδιαίτερα χαρακτηριστικά που παρουσιάζουν οι λαοί της Νοτιοανατολικής Ευρώπης επηρεάζουν την στάση τους απέναντι στους στόχους της “Πράσινης Συμφωνίας”. Επίσης, ενδιαφέρον είναι να δούμε πώς και αν τα ποσοστά των Βαλκανικών χωρών που είναι κράτη-μέλη της ΕΕ διαφέρουν από τα ποσοστά των Βαλκανικών χωρών που δεν είναι. Ακόμα, είναι ενδιαφέρον να δούμε τα νούμερα των Βαλκανίων, γιατί οι χώρες αυτές δεν έχουν βαριά βιομηχανία, όπως οι χώρες της Κεντρικής και Βόρειας Ευρώπης, ενώ ακόμα βασίζονται έντονα στις αγροτικές δραστηριότητες λόγω της μορφολογίας του εδάφους. Παρόλα αυτά, εξακολουθούν να είναι πηγές εκπομπών αερίων του θερμοκηπίου, εξαιτίας της παραγωγής ενέργειας από ορυκτά καύσιμα. Για αυτό τον λόγο, η έρευνα αφορά τις εκπομπές αερίων του θερμοκηπίου και παράλληλα την παραγωγή και κατανάλωση πράσινης ενέργειας.
